```python
import pickle
from pathlib import Path
import pandas as pd
```

```python
file_path = Path("output/GROUPE_DELICE/DELICE_tables.pkl")

with open(file_path, "rb") as f:
    data = pickle.load(f)

print("Type de données :", type(data))
```

```python
print("Contenu (clés principales) :")
print(data.keys())
```

```python
for year, tables in data.items():
    print("\n" + "="*60)
    print(f"📅 Année : {year}")
    print(f"📊 Nombre de tableaux : {len(tables)}")

    for i, df in enumerate(tables):
        print("\n------------------------------")
        print(f"Tableau {i+1}")

-----------------------------   display(df)   # Jupyter affiche le DataFrame proprement
```

```python
year = 2014  # 👉 change ici l'année que tu veux

if year in data:
    tables = data[year]

    print("="*60)
    print(f"📅 Année : {year}")
    print(f"📊 Nombre de tableaux : {len(tables)}")

    for i, df in enumerate(tables):
        print("\n------------------------------")
        print(f"Tableau {i+1}")
        display(df)
else:
    print("❌ Année introuvable :", year)
```
