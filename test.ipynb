```python
import pickle
import pandas as pd
```

```python
# Chemin du fichier .pkl
file_path = "output/GROUPE_DELICE/DELICE_tables.pkl"

# Charger les données
with open(file_path, "rb") as f:
    data = pickle.load(f)

# Afficher tous les tableaux de toutes les années
for year, tables in data.items():
    print("\n" + "=" * 60)
    print(f"📅 Année : {year} | {len(tables)} tableau(x)")
    print("=" * 60)

    for i, df in enumerate(tables, start=1):
        print(f"\n--- Tableau {i} ---")
        display(df)
```

```python
for year, tables in data.items():
    for i, df in enumerate(tables):
        print(f"Year {year} - Table {i}")
        print(df.head(3))  # juste aperçu
```

```python
all_df = []

for year, tables in data.items():
    for df in tables:
        df = df.copy()
        df["year"] = year
        all_df.append(df)

final_df = pd.concat(all_df, ignore_index=True)
print(final_df.shape)
print(final_df.empty)
```
