```python
import pickle
from pathlib import Path
import pandas as pd
```

```python
file_path = Path("output/GROUPE_DELICE/DELICE_tables.pkl")

with open(file_path, "rb") as f:
    data = pickle.load(f)

print("Type de données :", type(data))
```

```python
print("Contenu (clés principales) :")
print(data.keys())
```

```python
# Supposons que 'data' soit ton dict {year: [tables]}
year = 2014   # année cible

if year in data:
    tables = data[year]
    print("\n" + "="*60)
    print(f"📅 Année : {year}")
    print(f"📊 Nombre de tableaux : {len(tables)}")

    for i, df in enumerate(tables):
        print("\n------------------------------")
        print(f"Tableau {i+1}")
        display(df)   # Jupyter affiche le DataFrame proprement
else:
    print(f"⚠️ Aucun tableau trouvé pour l'année {year}")
```
